```python
import numpy as np
#Mean bases anomaly detection

x = pd.Series([2.1,2.3,4.6,2.0,2.5])
median = np.median(x)
treshold = 2
outliers = []
for item in x:
    if abs(item-median) > treshold:
        outliers.append(item)
print(outliers)
```

```python
#Z-Score-Based Anomaly Detection
x = pd.Series([2.1,2.3,4.6,2.0,2.5])

mean = np.mean(x)
std = np.std(x)
outliers = []
for item in x:
    z_score = (item - mean) / std
  #  print(z_score)
    if z_score > 1.5:
        outliers.append(item)
        print(z_score)


print(outliers)
```

```python
#Interquartile Range for Anomaly Detection
x = pd.Series([2.1, 2.3, 4.5, 2.2, 2.5])

Q1, Q3 = np.percentile(x,[25,75])
IQR = Q3 - Q1
outliers = []
for item in x:
    if item < (Q1 - 1.5 *IQR) or item > (Q3 + 1.5* IQR):
        outliers.append(item)

print(outliers)
```
