```python
x = pd.Series([1,2,3,4,5,6,7,150])
mean = x.mean()
std = x.std()
z_scores = abs((x-mean)/std)
z_scores
```

```python
outliers_removed = x[z_scores <= 1.5]
outliers_removed
```

```python
x = pd.Series([1,2,3,4,5,6,7,150])
mean = x.mean()
median = np.median(x)
z_scores = abs((x-mean)/std)
median
```

```python
x[z_scores > 1.5]  =median
x
```
